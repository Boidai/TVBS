```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver import Chrome
import time, re
import requests
import os
from selenium import webdriver
import json
import sys

def store(data):
    with open('TVBS.json','a',encoding='utf8') as f:
        f.write(data.encode(sys.stdin.encoding, "replace").decode(sys.stdin.encoding))

driver = Chrome("./chromedriver")
driver.get("https://news.tvbs.com.tw/politics/")
count = 0

for i in range(1,100):#滾動次數
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    time.sleep(1)

for page in range(0, 100):
    try:
        page = page + 1
        web = "#block_768 > li:nth-child("+str(page)+")"
        driver.find_element_by_css_selector(str(web)).click()
        url = driver.current_url
        print(url)
        response = urlopen(str(url))
        html = BeautifulSoup(response)
        news = html.find_all("div", {"class":"content"})
    
    except:
        continue
        
    for r in news:
        try:
            title = r.find("h1", class_="margin_b20").text 
            content = r.find("div", class_="h7 margin_b20").text
            date = r.find("div", class_="icon_time time leftBox2").text 
            author = r.find("h4", class_="font_color5 leftBox1").text
            kw = r.find("h5").text
            print(title, url, date, author, kw, content)
            count += 1
            print('第',count)
        
            newsd={"source":"TVBS","title":title,"url":url,"date_":date,"type":"net",
                "author":author,"kw":kw,"views":None,"content":content}
        
            json_data = json.dumps(newsd, ensure_ascii=False, indent=4, sort_keys=True)
            store(json_data)
        
            driver.back()
        
            time.sleep(5)
        
        except:
            continue
            
driver.close()            
            
print("Yeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeeee")    
```
