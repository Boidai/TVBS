```python
import requests
import os
from bs4 import BeautifulSoup
from urllib.request import urlopen
from selenium import webdriver
import time, re
from collections import defaultdict, OrderedDict
import sys
import json
import pandas as pd
import numpy as np
from urllib import request


def store(data):
    with open('2018-12-20 event.json','a',encoding='utf8') as f:
        f.write(data.encode(sys.stdin.encoding, "replace").decode(sys.stdin.encoding))

count = 0
nlist = []    

driver = webdriver.Chrome()
driver.implicitly_wait(30)
page = 1

while page<181:
    driver.get("https://news.tvbs.com.tw/news/searchresult/news/"+str(page)+"/?search_text=%E9%87%91%E9%A6%AC")
    try:
        sp = BeautifulSoup(driver.page_source)
    except HTTPError:
        print("大概是結束了")
        break
    for j in range(0,2000):#總共多少,設一個較大的數字 有多少篇數
        try:
            first = sp.find_all("div", {"class":"search_list_div"})[0]
            two = first.find("ul")
            three = two.find_all('li')[j]#變量 0~拉到底
            #four = three.find_all("div", {"class":"content_center_contxt_real_news"})[0]

        except:
            continue
        for i in range(0,1):#一次抓取數量
            try:
                four = three.find_all("a", {"class":"search_list_box"})[0]
                five = four['href']
                news_url = five

                title = four.find("div", {"class":"search_list_txt"}).text

                html1 = requests.get(news_url)
                html1.encoding="utf-8"

                sp1 = BeautifulSoup(html1.text, 'html.parser')

                ct = sp1.find("div", {"class":"h7 margin_b20"}).text
                ct1 = ct.replace("\t", "").replace("   ","").replace("\n", "").replace("（中央社）","")
                ct2 = ct1.replace("最HOT話題在這！想跟上時事，快點我加入TVBS新聞LINE好友！ ","")
                ct3 = ct2.replace("最HOT話題在這！想跟上時事，","").replace("快點我加入TVBS新聞LINE好友！ ","")
                ct4 = ct3.replace("TVBS最新大數據分析","").replace("看完整內容快點我加入TVBS新聞LINE好友！","")


                da = sp1.find("div", {"class":"icon_time time leftBox2"}).text

                au = sp1.find("h4", {"class":"font_color5 leftBox1"}).text
                au1 = au.replace("編輯  ","").replace(" 報導","").replace("記者","").replace("/ 攝影","").replace("  ","").replace(" ",",")
                au2 = au1.split(",")

                kw = sp1.find("div", {"class":"adWords"}).text
                kw1 = kw.replace("\t", "")
                kw2 = kw1.replace(",","")
                kw3 = kw2.replace("\n", "").replace("編輯  ","").replace(" 報導","")
                kw4 = kw3.replace("記者","").replace("      ",",")
                kw5 = kw4.split(",")

                img = sp1.find("div", {"class":"img margin_b20"})
                img1 = img.select('img')[0]['src']

                print(title, news_url, da, au1, kw4, ct4, img1)
                count += 1
                print('第',count)


                newsd = {"source":"TVBS","title":title,"url":news_url,"date_":da,
                 "type":"net","author":au2,"kw":kw5,"views":None,"content":ct4,
                    "img":img1,"event":"goldenhorse"}

                nlist.append(newsd)


            except:
                continue
    page = page + 1
                        
json_data = json.dumps(nlist, ensure_ascii=False, indent=4, sort_keys=True)
store(json_data)

print("下載完了!")

driver.close()
```
